```python

from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
data["Sex"]=label_encoder.fit_transform(data["Sex"])
data["ChestPainType"]=label_encoder.fit_transform(data["ChestPainType"])
data["RestingECG"]=label_encoder.fit_transform(data["RestingECG"])
data["ExerciseAngina"]=label_encoder.fit_transform(data["ExerciseAngina"])
data["ST_Slope"]=label_encoder.fit_transform(data["ST_Slope"])
data
```

```python
data
```

```python
x=data[["Age","Sex","ChestPainType","RestingBP","Cholesterol","FastingBS","RestingECG","MaxHR","ExerciseAngina","Oldpeak","ST_Slope"]]
print(x)
```

```python
y=data[["HeartDisease"]]
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train, y_train)
```

```python
x_test
```

```python
y_test
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
import joblib
joblib.dump(model,"heartdisease1.pk1")

```
